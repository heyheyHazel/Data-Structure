## 2.1 æ ˆçš„å®šä¹‰
æ ˆæ˜¯ä¸€ç§**çº¿æ€§ç»“æ„**ï¼ŒåŠ å…¥å’Œç§»é™¤éƒ½å‘ç”Ÿåœ¨**åŒä¸€ç«¯**ï¼Œç¬¦åˆ**åè¿›å…ˆå‡º**åŸåˆ™ï¼ŒLIFO(Last in First out)ã€‚ä¾‹å¦‚ï¼šä¸€å †ä¹¦ç±ğŸ“šï¼Œæœ€è¿‘è®¿é—®çš„ç½‘ç«™Â·Â·Â·

```python
class Stack1:
    '''æ ˆé¡¶åœ¨å°¾ç«¯ å¤æ‚åº¦O(1)'''
    def __init__(self):
        self.items = [] # åˆ›å»ºä¸€ä¸ªç©ºçš„æ ˆï¼ˆåˆ—è¡¨ï¼‰

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.isEmpty():
            return self.items.pop()
        raise IndexError("pop from empty stack")

    def peek(self):
        if not self.isEmpty():
            return self.items[-1]
        raise IndexError("peek from empty stack")

    def isEmpty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
    
    def getItems(self):
        return self.items

# ç¤ºä¾‹ç”¨æ³•
if __name__ == "__main__":
    stack = Stack1()
    stack.push(1)   # åŠ å…¥å…ƒç´ 1
    stack.push(2)   # åŠ å…¥å…ƒç´ 2
    stack.push(3)   # åŠ å…¥å…ƒç´ 3
    print("æ ˆå†…å®¹:", stack.getItems())  # è¾“å‡º: æ ˆå†…å®¹: [1, 2, 3]
    print("æ ˆé¡¶å…ƒç´ :", stack.peek())  # è¾“å‡º: æ ˆé¡¶å…ƒç´ : 3
    print("æ ˆçš„å¤§å°:", stack.size())   # è¾“å‡º: æ ˆçš„å¤§å°: 3
    print("å¼¹å‡ºå…ƒç´ :", stack.pop())   # è¾“å‡º: å¼¹å‡ºå…ƒç´ : 3
    print("æ ˆæ˜¯å¦ä¸ºç©º:", stack.isEmpty()) # è¾“å‡º: æ ˆæ˜¯å¦ä¸ºç©º: False
```

```python
class Stack2():
    '''æ ˆé¡¶åœ¨å¤´ç«¯ å¤æ‚åº¦O(n)'''
    def __init__(self):
        self.items = [] # åˆ›å»ºä¸€ä¸ªç©ºçš„æ ˆï¼ˆåˆ—è¡¨ï¼‰

    def push(self, item):
        self.items.insert(0, item)

    def pop(self):
        if not self.isEmpty():
            return self.items.pop(0)
        raise IndexError("pop from empty stack")

    def peek(self):
        if not self.isEmpty():
            return self.items[0]
        raise IndexError("peek from empty stack")

    def isEmpty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def getItems(self):
        return self.items

# ç¤ºä¾‹ç”¨æ³•
if __name__ == "__main__":  
    stack = Stack2()
    stack.push(1)   # åŠ å…¥å…ƒç´ 1
    stack.push(2)   # åŠ å…¥å…ƒç´ 2
    stack.push(3)   # åŠ å…¥å…ƒç´ 3
    print("æ ˆå…ƒç´ :", stack.getItems())
    print("æ ˆé¡¶å…ƒç´ :", stack.peek())  # è¾“å‡º: æ ˆé¡¶å…ƒç´ : 3
    print("æ ˆçš„å¤§å°:", stack.size())   # è¾“å‡º: æ ˆçš„å¤§å°: 3
    print("å¼¹å‡ºå…ƒç´ :", stack.pop())   # è¾“å‡º: å¼¹å‡ºå…ƒç´ : 3
    print("æ ˆæ˜¯å¦ä¸ºç©º:", stack.isEmpty()) # è¾“å‡º: æ ˆæ˜¯å¦ä¸ºç©º: False
```

```python
def parChecker(symbolString):
    s = Stack1()
    balanced = True
    index = 0

    while index < len(symbolString) and balanced:
        symbol = symbolString[index]
        if symbol == '(':
            s.push(symbol)
        if symbol == ')':
            if s.isEmpty():
                balanced = False
            else:
                s.pop()
        else:
            pass
        index += 1
    if balanced and s.isEmpty():
        return True
    else:
        return False

print(parChecker('(dd)'))
print(parChecker('(())))))'))
```

æ‹“å±•ä¸ºæ›´å¤šç§ç±»æ‹¬å·çš„åŒ¹é…ï¼š()[]{}ã€‚éœ€è¦è€ƒè™‘ï¼š1ï¼‰æ‹¬å·æ•°é‡çš„åŒ¹é…ï¼›2ï¼‰ä¸èƒ½å‡ºç°ä¸åŒç±»å‹ç©¿æ’çš„æƒ…å†µï¼Œä¾‹å¦‚([)]ã€‚

```python
def parChecker2(symbolString):
    s = Stack1()
    balanced = True
    index = 0
    while index < len(symbolString) and balanced:
        symbol = symbolString[index]
        if symbol in '([{':
            s.push(symbol)
        if symbol in ')]}':
            if s.isEmpty():
                balanced = False
            else:
                top = s.pop()
                if not matches(top, symbol):
                    balanced = False
        index += 1
    if balanced and s.isEmpty():
        return True
    else:   
        return False

def matches(open, close):
    opens = '([{'
    closes = ')]}'
    return opens.index(open) == closes.index(close) # é€šè¿‡æ‹¬å·çš„ç´¢å¼•ä½ç½®æ¥åˆ¤æ–­æ˜¯å¦åŒ¹é…

print(parChecker2('{[()]}'))
print(parChecker2('{[(])}'))
```

åè¿›åˆ¶è½¬æ¢ä¸ºåå…­ä»¥ä¸‹ä»»æ„è¿›åˆ¶çš„ä»£ç ï¼šå¤šå°‘è¿›åˆ¶å°±é™¤ä»¥å¤šå°‘ï¼ŒäºŒè¿›åˆ¶é™¤ä»¥äºŒï¼Œå…«è¿›åˆ¶é™¤ä»¥å…«ï¼ŒåŒæ ·æ±‚ä½™æ•°ã€‚

```python
def baseConverter(decNumber, base):
    digits = '0123456789ABCDEF'
    if base < 2 or base > 16:
        raise ValueError("Base must be between 2 and 16")
    
    s = Stack1()
    while decNumber > 0:
        rem = decNumber % base
        s.push(rem)
        decNumber = decNumber // base
    
    newString = ''
    while not s.isEmpty():
        newString = newString + digits[s.pop()]
    
    return newString

print(baseConverter(25, 8))
print(baseConverter(1307, 16))
```
